```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
import os
```

## Convection-diffusion equation

```python
Nx = Ny = 201;
loadname='.\\data\\solData_heat.txt'
u = np.loadtxt(loadname)
val = 1 * u.max()
img_num = int(u.size / Nx / Ny)
u = u.reshape(img_num, Nx, Ny);
fig, axes = plt.subplots(2,2, figsize=(8,8))
ind = 0
for ax in axes.flat:
    ax.imshow(u[int(img_num/5)*ind], cmap="inferno", vmin=0, vmax=val)
    ind += 1
fig.tight_layout()
```

```python
# save gif

filenames = []
savename='.\\data\\heat.gif'

i = 0
for img in u:
    plt.imshow(img, cmap="inferno", vmin=0, vmax=val)
    filename = f'{i}.png'
    filenames.append(filename)
    i += 1

    plt.savefig(filename, dpi=100)
    plt.close()
# build gif
with imageio.get_writer(savename, mode='I') as writer:
    for filename in filenames:
        image = imageio.v3.imread(filename)
        writer.append_data(image)
        
# Remove files
for filename in set(filenames):
    os.remove(filename)
```

## Maxwell's equations in 2D, transverse electric (TE) mode

```python
Nx = Ny = 201;
loadname='.\\data\\solData_wave.txt'
u = np.loadtxt(loadname)
val = 0.2 * u.max()
img_num = int(u.size / Nx / Ny)
u = u.reshape(img_num, Nx, Ny);
fig, axes = plt.subplots(2,2, figsize=(8,8))
ind = 1
for ax in axes.flat:
    ax.imshow(u[int(img_num/5)*ind], cmap="bwr", vmin=-val, vmax=val)
    ind += 1
fig.tight_layout()
```

```python
# save gif
filenames = []
savename='.\\data\\wave.gif'
i = 0
for img in u:
    plt.imshow(img, cmap="bwr", vmin=-val, vmax=val)
    filename = f'{i}.png'
    filenames.append(filename)
    i += 1

    plt.savefig(filename, dpi=100)
    plt.close()
# build gif
with imageio.get_writer(savename, mode='I') as writer:
    for filename in filenames:
        image = imageio.v3.imread(filename)
        writer.append_data(image)
        
# Remove files
for filename in set(filenames):
    os.remove(filename)
```
