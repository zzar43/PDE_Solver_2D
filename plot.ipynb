```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = np.loadtxt('.\\data\\solData.txt')
val = 0.2 * u.max()
# val = 1e-1
u = u.reshape(11, 201, 201);
fig, axes = plt.subplots(2,2)
ind = 2
for ax in axes.flat:
    ax.imshow(u[2*ind], cmap="bwr", vmin=-val, vmax=val)
    ind += 1
fig.tight_layout()
```
